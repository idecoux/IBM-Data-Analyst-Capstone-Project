## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
df = pd.read_sql_query("SELECT * FROM master", conn)
df

import matplotlib.pyplot as plot

df = pd.read_sql_query("SELECT * FROM master", conn)
plot.hist(df['ConvertedComp'])

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df

df['Age'].plot(kind='box', figsize=(6, 10))
plot.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age , WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df

df.plot(kind='scatter', x = 'Age' , y= 'WorkWeekHrs' )
plot.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age , WorkWeekHrs , CodeRevHrs
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df

ax = df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s = 'Age', figsize=(14, 10), alpha=0.5,  color='green')

plot.title('Age Bubbles on Code Revision Hours and Work Week Hours')
plot.xlabel('WorkWeekHrs')
plot.ylabel('CodeRevHrs')
plot.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """

SELECT count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count(DatabaseDesireNextYear)
DESC LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

df['Count'].plot(kind='pie',figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,\
                            labels=df['DatabaseDesireNextYear'])

plot.title('Top 5 databases that respondents wish to learn next year')
plot.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE age >=30 AND age<=35
"""
df_age = pd.read_sql_query(QUERY,conn) 

df_age = df_age.groupby('Age', axis=0).median()

df_age.plot(kind='bar', figsize=(10, 6), stacked=True)

plot.xlabel('Age')
plot.ylabel('Hours') 
plot.title('Median Hours by Age')

plot.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE age >=45 AND age<=60
"""
df_comp_age = pd.read_sql_query(QUERY,conn) 

df_comp_age = df_comp_age.groupby('Age', axis=0).median()


df_comp_age.plot(kind='line', figsize=(10, 6), stacked=True)

plot.xlabel('Age')
plot.ylabel('$')
plot.title('Median Compensation by Age')

plot.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, count(MainBranch) as Count
FROM master
GROUP BY MainBranch
"""
df_mainbranch = pd.read_sql_query(QUERY,conn)
print (df_mainbranch)

df_mainbranch.plot(kind='barh', figsize=(10, 6))

plot.xlabel('Number of Respondents')
plot.ylabel('Main Branch')
plot.title('Number of Respondents by Main Branch')

plot.show()
```
